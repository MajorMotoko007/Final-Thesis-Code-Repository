```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# Step 1: Load the CSV file
df = pd.read_csv('export2025.06.13-12.13.39.csv')  # Ensure file is in the same folder as the notebook

# Step 2: Extract authors from 'Authors' column
author_list = []
for entry in df['Authors'].dropna():
    names = [name.strip() for name in entry.split(';') if name.strip()]
    author_list.extend(names)

# Step 3: Count all author frequencies
author_counts = Counter(author_list)
top_authors = author_counts.most_common(10)
# Step 4: Convert to sorted lists
names, counts = zip(*top_authors)

# Step 5: Plot horizontal bar chart of all authors
plt.figure(figsize=(10, 6))
plt.barh(names, counts, color='steelblue')
plt.xlabel('Document Count')
plt.title('Top Authors by Frequency')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
